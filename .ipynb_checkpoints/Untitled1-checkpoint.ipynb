```python
import numpy as np
import matplotlib.pyplot as plt
file=open("HR data.txt", 'r')
wholefile=file.readlines()
V_mag=np.empty(2720)
BV_mag=np.empty(2720)
j=0
for i in range (2,2720):
    strreader=wholefile[i]
    elements=strreader.split()
#     elements=np.array(elements)
    if len(elements)==9:
        V_mag[j]=elements[1]
        BV_mag[j]=elements[8]
        j+=1


    
    
```

```python
# for i in range(2720):
#     print(V_mag[i],BV_mag[i])
```

```python
plt.scatter( BV_mag,V_mag,marker='.', color='blue')

plt.xlabel('BV_mag')
plt.ylabel('V_mag')
plt.title('HR diagram')
plt.gca().invert_yaxis()
plt.show()
    
```
