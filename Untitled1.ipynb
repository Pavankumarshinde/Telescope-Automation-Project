```python
import numpy as np
import matplotlib.pyplot as plt

with open("applied_ststs_data.csv", 'r') as file:
    wholefile = file.readlines()
Array = np.zeros(8)

for i in range(1, len(wholefile)-2):  # Start from 1 to skip the header
    elements = wholefile[i].split(',')
    
    runs = int(elements[5])
    print(runs)
    # if(runs<100):
    #     print("true")
    

    if 0 < runs and runs < 100:
        Array[0] += 1
    elif 100 < runs and runs < 200:
        Array[1] += 1
    elif 200 < runs and runs < 300:
        Array[2] += 1
    elif 300 < runs and runs < 400:
        Array[3] += 1
    elif 400 < runs and runs < 500:
        Array[4] += 1
    elif 500 < runs and runs < 600:
        Array[5] += 1
    elif 600 < runs and runs < 700:
        Array[6] += 1
    elif 700 < runs:
        Array[7] += 1

bin_edges = ['0-100', '100-200', '200-300', '300-400', '400-500', '500-600', '600-700', '700-800']  # Define custom bin edges
plt.bar(bin_edges, Array, align='center')  # Plot histogram

plt.xlabel('Runs')
plt.ylabel('frequency')
plt.title('Runs Distribution')
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the data into a DataFrame
file_path = "RA_DE_data.txt"
df = pd.read_csv(file_path, delim_whitespace=True)

# Extract the RA and DE columns
RA = df["RA"]
DE = df["DE"]

# Convert RA and DE to radians
RA_radians = np.deg2rad(RA)  # Assuming RA is already in degrees
DE_radians = np.deg2rad(DE)

# Adjust RA to be within the range -180 to 180 degrees in radians
RA_radians = np.remainder(RA_radians, 2*np.pi)  # RA should be in [0, 2pi]
RA_radians[RA_radians > np.pi] -= 2*np.pi  # Shift RA from [0, 2pi] to [-pi, pi]

# Create a Mollweide projection plot
plt.figure(figsize=(10, 6))
ax = plt.subplot(111, projection="mollweide")

# Plot the data
sc = ax.scatter(RA_radians, DE_radians, s=10, c='blue', alpha=0.5)
ax.set_title("Mollweide Projection of RA vs DE")
ax.grid(True)

# Display the plot
plt.show()
```
