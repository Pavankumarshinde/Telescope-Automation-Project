```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("C:\\Users\\Sudha\\Downloads\\hipparcos.csv")
df.head()
```

```python
ra = df['RA']
de = df['DE']
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 5))
plt.scatter(ra, de, s=0.1, color='black')
plt.xlabel('Right Ascension (degrees)')
plt.ylabel('Declination (degrees)')
plt.title('RA-DEC Sky Plot for HIPPARCOS Catalog')
plt.grid(True)
plt.show()
```
