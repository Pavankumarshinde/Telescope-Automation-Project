```python
import turtle
import math

# Constants
G = 6.67430e-11  # Gravitational constant (m^3 kg^(-1) s^(-2))
m1 = 5.972e24    # Mass of Earth (kg)
m2 = 7.3477e22   # Mass of Moon (kg)
initial_distance = 384400000  # Initial distance between Earth and Moon (m)
initial_velocity = 1022  # Initial velocity of Moon (m/s)
time_step = 1000  # Time step for simulation (seconds)

# Create the screen
screen = turtle.Screen()
# screen.bgcolor("")

# Create the bodies
earth = turtle.Turtle()
earth.shape("circle")
earth.color("blue")
earth.shapesize(1)  # Adjust the size of Earth

moon = turtle.Turtle()
moon.shape("circle")
moon.color("gray")
moon.shapesize(0.5)  # Adjust the size of the Moon

# Initialize positions and velocities
earth.penup()
earth.goto(0, 0)
earth.pendown()

moon.penup()
moon.goto(0, -initial_distance)
moon.pendown()

moon_velocity_x = initial_velocity  # Initial velocity of Moon along the x-axis
moon_velocity_y = 0  # Initial velocity of Moon along the y-axis

# Simulation loop
for _ in range(5000):
    # Calculate the distance between Earth and Moon
    x_distance = moon.xcor()
    y_distance = moon.ycor()
    distance = math.sqrt(x_distance**2 + y_distance**2)


    # Calculate the gravitational force
    force = G * m1 * m2 / distance**2

    # Calculate the acceleration
    acceleration_x = (force / m2) * (x_distance / distance)
    acceleration_y = (force / m2) * (y_distance / distance)

    # Update the velocities
    moon_velocity_x += acceleration_x * time_step
    moon_velocity_y += acceleration_y * time_step

    # Update the positions
    moon.goto(moon.xcor() + moon_velocity_x * time_step, moon.ycor() + moon_velocity_y * time_step)

# Close the simulation when clicking on the screen
screen.exitonclick()
```

```python
from turtle import Turtle, Screen
import math

# Masses of the bodies
m1 = 30
m2 = 40
m3 = 50

# Gravitational constant
G = 100

# Create the screen
screen = Screen()
screen.bgcolor("white")
screen.setup(width=800, height=800)
screen.title("Three Body Problem")

# Create the bodies
body1 = Turtle(shape='circle')
body2 = Turtle(shape='circle')
body3 = Turtle(shape='circle')

body1.shapesize(m1 / 10)  # Set the size of body1 based on mass
body2.shapesize(m2 / 10)
body3.shapesize(m3 / 10)

body1.penup()
body2.penup()
body3.penup()

body1.color("blue")
body2.color("red")
body3.color("yellow")

body1.goto(-50, -50)
body2.goto(50, 50)
body3.goto(-50, 50)

# Initialize velocities
body1_dx = 10
body1_dy = 10
body2_dx = 10
body2_dy = 10
body3_dx = 10
body3_dy = 10

# Function to calculate the force between two bodies
def calculate_force(m1, m2, r):
    force = (G * m1 * m2) / (r ** 2)
    return force

# Function to update positions and velocities
def move():
    global body1_dx, body1_dy, body2_dx, body2_dy, body3_dx, body3_dy

    r1 = math.sqrt((body2.xcor() - body1.xcor()) ** 2 + (body2.ycor() - body1.ycor()) ** 2)
    r2 = math.sqrt((body3.xcor() - body2.xcor()) ** 2 + (body3.ycor() - body2.ycor()) ** 2)
    r3 = math.sqrt((body3.xcor() - body1.xcor()) ** 2 + (body3.ycor() - body1.ycor()) ** 2)

    F1 = calculate_force(m1, m2, r1)
    F2 = calculate_force(m3, m2, r2)
    F3 = calculate_force(m1, m3, r3)

    # Calculate accelerations
    a1 = F1 / m1
    a2 = F2 / m2
    a3 = F3 / m3

    # Update velocities
    body1_dx += a1
    body1_dy += a1
    body2_dx += a2
    body2_dy += a2
    body3_dx += a3
    body3_dy += a3

    # Update positions
    body1.goto(body1.xcor() + body1_dx, body1.ycor() + body1_dy)
    body2.goto(body2.xcor() + body2_dx, body2.ycor() + body2_dy)
    body3.goto(body3.xcor() + body3_dx, body3.ycor() + body3_dy)

    screen.update()
    screen.ontimer(move, 100)

# Start the animation
move()

screen.listen()
screen.exitonclick()
```

```python
from turtle import Turtle, Screen

# Masse
m1 = 3.0e+10
m2 = 4.0e+10

# Gravitational constant
G = 6.67430e-11  # Actual value of the gravitational constant

# Time step
del_t = 0.1  # Time step for the simulation

# Create Turtle objects for the bodies
body1 = Turtle(shape='circle')
body2 = Turtle(shape='circle')

# Set the sizes and colors of the bodies
body1.shapesize(m1 / 10)
body2.shapesize(m2 / 10)
body1.color("blue")
body2.color("red")

# Set the initial positions of the bodies
body1.penup()
body2.penup()
body1.goto(-50, -50)
body2.goto(50, 50)

# Set the initial velocities for the bodies
u1x = 0  # Initial velocity for body 1 in the x-direction
u1y = 0  # Initial velocity for body 1 in the y-direction
u2x = 0  # Initial velocity for body 2 in the x-direction
u2y = 0  # Initial velocity for body 2 in the y-direction

# Initialize simulation variables
gameon = True
i = 0

# Define the move function to update positions and velocities
def move():
    global gameon, i, u1x, u1y, u2x, u2y
    
    # Calculate the distance between the bodies
    r1 = body1.distance(body2)
    
    # Calculate gravitational forces
    F1 = G * m1 * m2 / (r1 ** 2)
    Fx1 = F1 * ((body2.xcor() - body1.xcor()) / r1)
    Fy1 = F1 * ((body2.ycor() - body1.ycor()) / r1)
    Fx2 = -Fx1
    Fy2 = -Fy1
    
    # Update velocities
    u1x += del_t * Fx1 / m1
    u1y += del_t * Fy1 / m1
    u2x += del_t * Fx2 / m2
    u2y += del_t * Fy2 / m2
    
    # Update positions
    body1.goto(body1.xcor() + u1x * del_t, body1.ycor() + u1y * del_t)
    body2.goto(body2.xcor() + u2x * del_t, body2.ycor() + u2y * del_t)
    
    # Increment the time step counter
    i += 1
    
    # Print the current time step
    print("Time Step:", i)
    
    # Terminate the simulation after a certain number of time steps
    if i > 1000:
        gameon = False

# Run the simulation
while gameon:
    move()

# Create the screen and set it up
screen = Screen()
screen.setup(width=800, height=800)
screen.title("Two Body Problem")
screen.exitonclick()
```
